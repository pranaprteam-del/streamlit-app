# Get Full Summary column

```python
import requests
from bs4 import BeautifulSoup

def extract_full_text_from_span_container(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                      "(KHTML, like Gecko) Chrome/115.0 Safari/537.36"
    }
    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code != 200:
            return f"Failed to fetch page, status code: {response.status_code}"

        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the target span by ID
        span = soup.find('span', id='ctl00_ContentPlaceHolder1_TabContainer1_TabPanel2_lblNews')
        if not span:
            return "Target span not found"

        # Get the parent div container of the span (adjust as needed)
        container_div = span.find_parent('div')
        if not container_div:
            # fallback to span text only
            return span.get_text(separator=' ', strip=True)

        # Extract all text inside this div, including all child spans, divs, etc.
        full_text = container_div.get_text(separator=' ', strip=True)

        return full_text

    except Exception as e:
        return f"Error occurred: {str(e)}"


df['full_summary'] = df.apply(
    lambda row: extract_full_text_from_span_container(row['url']) 
                if pd.notna(row['Summary']) and row['Summary'].strip() != "" 
                else None,
    axis=1
)
```

# Update Summary column for regional text using image url

```python
import pandas as pd
import pytesseract
from PIL import Image
import requests
from io import BytesIO
from bs4 import BeautifulSoup
from urllib.parse import urljoin

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# --- Language Mapping: DataFrame value -> Tesseract language code ---
LANGUAGE_MAP = {
    'english': 'eng', 'en': 'eng',
    'hindi': 'hin', 'hi': 'hin',
    'bengali': 'ben', 'bangla': 'ben', 'bn': 'ben',
    'telugu': 'tel', 'te': 'tel',
    'tamil': 'tam', 'ta': 'tam',
    'marathi': 'mar', 'mr': 'mar',
    'gujarati': 'guj', 'gu': 'guj',
    'kannada': 'kan', 'kn': 'kan',
    'malayalam': 'mal', 'ml': 'mal',
    'punjabi': 'pan', 'pa': 'pan',
    'odia': 'ori', 'oriya': 'ori', 'or': 'ori',
    'assamese': 'asm', 'as': 'asm',
    'sanskrit': 'san', 'sa': 'san'
}

TESS_LANGS_OF_INTEREST = set(LANGUAGE_MAP.values())

# --- Detect language from image using sequential try ---
def detect_language_simple(image_pil):
    languages_to_try = ['hin', 'ben', 'tel', 'tam', 'mar', 'guj', 'kan', 'eng']
    best_lang = 'eng'
    max_alpha_ratio = 0.0

    for lang in languages_to_try:
        try:
            text = pytesseract.image_to_string(image_pil, lang=lang, config='--psm 6').strip()
            if not text:
                continue

            alpha_count = sum(1 for c in text if c.isalpha())
            total_chars = len(text)
            if total_chars == 0:
                continue

            alpha_ratio = alpha_count / total_chars
            if alpha_ratio > max_alpha_ratio:
                best_lang = lang
                max_alpha_ratio = alpha_ratio
        except Exception as e:
            print(f"Error with lang '{lang}': {e}")
            continue

    print(f"Detected language: {best_lang} (confidence: {max_alpha_ratio:.2f})")
    return best_lang

# --- Determine Tesseract language code to use ---
def get_tess_lang_code(df_language_value, image_pil):
    if pd.notna(df_language_value) and str(df_language_value).strip() != "":
        df_lang_key = str(df_language_value).strip().lower()
        tess_code = LANGUAGE_MAP.get(df_lang_key)
        if tess_code:
            print(f"Using DataFrame language: {df_lang_key} -> {tess_code}")
            return tess_code
        else:
            print(f"Warning: Language '{df_language_value}' not in LANGUAGE_MAP. Falling back to detection.")

    print("No valid language in DataFrame. Detecting language from image...")
    return detect_language_simple(image_pil)

# --- Extract the actual image URL from the page ---
def get_image_url(page_url, img_id):
    try:
        response = requests.get(page_url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        
        img_tag = soup.find('img', id=img_id)
        if img_tag and img_tag.get('src'):
            return urljoin(page_url, img_tag['src'])
        else:
            print(f"No image found with id '{img_id}' in {page_url}")
            return None
    except Exception as e:
        print(f"Failed to fetch image URL from {page_url}: {e}")
        return None

# --- OCR from Image URL ---
def extract_text_from_url(image_url, df_language_value):
    try:
        response = requests.get(image_url)
        response.raise_for_status()

        # Check if the response is actually an image
        if 'image' not in response.headers.get('Content-Type', ''):
            raise ValueError("URL did not return an image")

        image = Image.open(BytesIO(response.content))

        lang_code = get_tess_lang_code(df_language_value, image)
        text = pytesseract.image_to_string(image, lang=lang_code, config='--psm 6')

        return ' '.join(text.splitlines())
    except Exception as e:
        print(f"Error processing {image_url}: {e}")
        return ""

# --- Load your DataFrame (replace this with actual CSV or data source) ---
# df = pd.read_csv("your_input_file.csv")  # Uncomment if loading from file

# --- Main processing loop ---
for idx, row in df.iterrows():
    if pd.isna(row['full_summary']) or str(row['full_summary']).strip() == "":
        page_url = row['url']
        img_url = get_image_url(page_url, img_id="ctl00_ContentPlaceHolder1_TabContainer1_TabPanel1_Image1")
        if img_url:
            extracted_text = extract_text_from_url(img_url, row.get('Language', ''))
            df.at[idx, 'full_summary'] = extracted_text
        else:
            df.at[idx, 'full_summary'] = ""

# --- Optional: Save the updated DataFrame ---
# df.to_csv("updated_file.csv", index=False)
```

# translate regional Summary text to english

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
import re

# Step 1: Set up the Chrome driver in headless mode (background)
def setup_driver():
    options = Options()
    options.add_argument("--disable-blink-features=AutomationControlled")
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option('useAutomationExtension', False)
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--no-sandbox")
    options.add_argument("--headless=new")  # ‚úÖ Headless mode
    options.add_argument("--window-size=1920,1080")  # Required for proper headless rendering
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")

    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=options)

    # Bypass automation detection
    driver.execute_cdp_cmd('Page.addScriptToEvaluateOnNewDocument', {
        'source': 'Object.defineProperty(navigator, "webdriver", {get: () => undefined})'
    })

    return driver

# Step 2: Open Google Translate once
def open_google_translate(driver):
    driver.get("https://translate.google.co.in/?hl=en&tab=TT&sl=auto&tl=en&op=translate")
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "textarea"))
    )
    time.sleep(2)  # Let page fully settle

# Step 3: Translate a single text (no page reload)
def translate_text_on_page(driver, text, max_retries=3):
    if not text or not text.strip():
        return text

    wait = WebDriverWait(driver, 15)
    for attempt in range(max_retries):
        try:
            textarea = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "textarea")))
            textarea.clear()
            textarea.send_keys(text)

            # ‚úÖ NEW: Wait 4 seconds before copying the translation
            time.sleep(4)

            # Try different selectors to find translated output
            translation_selectors = [
                "span[jsname='jqKxS']",
                "span[jsname='W297wb']",
                "span.ryNqvb"
            ]

            translated_text = ""
            for selector in translation_selectors:
                elements = driver.find_elements(By.CSS_SELECTOR, selector)
                translated_text = ' '.join([el.text for el in elements if el.text.strip()])
                if translated_text:
                    break

            if translated_text:
                return translated_text

            time.sleep(1)  # Wait before retry

        except Exception as e:
            print(f"Translation attempt {attempt + 1} failed: {e}")
            time.sleep(2)

    print(f"‚ö†Ô∏è Failed to translate text: {text[:50]}...")
    return text  # Return original if failed

# Step 4: Clean OCR or messy text
def clean_ocr_text(text):
    if pd.isna(text) or not isinstance(text, str):
        return ""
    return re.sub(r'\s+', ' ', text).strip()

# Step 5: Translate the DataFrame using the persistent driver and page
def translate_dataframe_with_open_driver(df, driver, text_column, language_column, output_column):
    if output_column not in df.columns:
        df[output_column] = ""

    mask = (df[language_column].str.lower() != 'english') & ((df[output_column].isna()) | (df[output_column] == ""))
    for idx in df[mask].index:
        text = df.at[idx, text_column]
        print(f"üîÑ Translating row {idx}...")
        translated = translate_text_on_page(driver, text)
        df.at[idx, output_column] = translated

    # Copy English text as-is
    english_mask = df[language_column].str.lower() == 'english'
    df.loc[english_mask, output_column] = df.loc[english_mask, text_column]

    return df

# Main script entry
if __name__ == "__main__":
    # üß™ Example DataFrame (replace with your actual dataset)
   
    # Step A: Clean the text
    df['summary_clean'] = df['full_summary'].apply(clean_ocr_text)

    # Step B: Set up the driver and translate
    print("üåê Starting headless translation session...")
    driver = setup_driver()
    try:
        open_google_translate(driver)
        df = translate_dataframe_with_open_driver(
            df,
            driver,
            text_column='summary_clean',
            language_column='Language',
            output_column='summary_trans'
        )
    finally:
        driver.quit()
        print("‚úÖ Chrome driver closed.")

    # Step C: Output results
    print("\nüìÑ Translated DataFrame:")
    df

    # # Step D: Save results to CSV
    # df.to_csv("translated_output.csv", index=False, encoding='utf-8')
    # print("\nüíæ Translation saved to 'translated_output.csv'")
```

```python
def clean_translation_output(text):
    if pd.isna(text) or not isinstance(text, str):
        return ""
    
    # Replace line breaks with space
    text = text.replace('\n', ' ')
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)

    # Strip leading/trailing space
    return text.strip()
df['summary_trans'] = df['summary_trans'].apply(clean_translation_output)
```

```python

import re
import unicodedata

def clean_text(text):
    """
    Cleans multilingual and repetitive/gibberish text while preserving meaningful English content.
    """

    # Normalize unicode to handle composed characters
    text = unicodedata.normalize("NFKC", text)

    # Remove repeated non-English character sequences (e.g. ‡¶ú ‡¶ú ‡¶ú ‡¶ú, ‡¨π‡¨æ‡¨è ‡¨π‡¨æ‡¨è ‡¨π‡¨æ‡¨è)
    text = re.sub(r'((\b[^\x00-\x7F\s]{1,3}\b)[\s]*){2,}', '', text)

    # Remove standalone non-Latin characters (single words)
    text = re.sub(r'\b[^\x00-\x7F]{1,4}\b', '', text)

    # Remove leftover multiple spaces
    text = re.sub(r'\s{2,}', ' ', text)

    # Remove any remaining junk characters or symbols
    text = re.sub(r'[^\w\s.,;:?!\'"-]', '', text)

    # Fix common sentence structure issues (e.g. double periods)
    text = re.sub(r'\.\.+', '.', text)

    # Optional: strip leading/trailing spaces
    text = text.strip()

    return text
    
df['summary_trans'] = df['summary_trans'].astype(str).apply(clean_text)
```

```python
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize

# Download required resources if not already present
nltk.download('punkt')
nltk.download('punkt_tab')

entity_dict = {
    'Mahindra': ['Mahindra', 'Club Mahindra', 'MHRIL', 'Mahindra Holidays']
}

def extract_entity_sentences(text, entity_dict):
    if not isinstance(text, str) or not text.strip():
        return ""
    
    # Use sent_tokenize instead of loading tokenizer directly
    sentences = sent_tokenize(text)
    all_entity_terms = [term.lower() for names in entity_dict.values() for term in names]
    
    matched_sentences = [sent for sent in sentences if any(term in sent.lower() for term in all_entity_terms)]
    
    return " ".join(matched_sentences).strip()

# Example DataFrame


# Apply extraction
df['short_summary'] = df['summary_trans'].apply(lambda x: extract_entity_sentences(x, entity_dict))

df['short_summary'][0]
```

```python
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from transformers import pipeline
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Load FinBERT model
finbert_tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
finbert_model = AutoModelForSequenceClassification.from_pretrained("yiyanghkust/finbert-tone")
finbert_pipeline = pipeline("sentiment-analysis", model=finbert_model, tokenizer=finbert_tokenizer)

# Initialize VADER
vader = SentimentIntensityAnalyzer()

def get_finbert_sentiment(text):
    result = finbert_pipeline(text)
    return result[0]['label']  # 'positive', 'neutral', or 'negative'

def get_vader_sentiment(text):
    scores = vader.polarity_scores(text)
    compound = scores['compound']
    if compound >= 0.05:
        return 'positive'
    elif compound <= -0.05:
        return 'negative'
    else:
        return 'neutral'

def hybrid_tonality(row):
    headline_sentiment = get_finbert_sentiment(row['Headline'])
    
    if headline_sentiment == 'negative':
        # Run VADER on full summary
        summary_sentiment = get_vader_sentiment(row['short_summary'])
        if summary_sentiment != 'negative':
            return summary_sentiment  # Override with VADER
        else:
            return headline_sentiment  # Keep FinBERT's 'negative'
    else:
        return headline_sentiment  # Keep FinBERT result

# Apply on your DataFrame
df['Tonality_Final'] = df.apply(hybrid_tonality, axis=1)

# Show results
df[['Headline', 'Tonality', 'Tonality_Final']]
```

```python
import spacy
import pandas as pd

# Load spacy model for NER
nlp = spacy.load("en_core_web_sm")

# Reuse your sentiment functions:
# get_finbert_sentiment(text), get_vader_sentiment(text), hybrid_tonality(text)

def get_entity_sentiments(text):
    doc = nlp(text)
    entity_sentiments = {}

    # Get sentences from the doc
    sentences = list(doc.sents)

    # Get unique entities (text)
    entities = set([ent.text for ent in doc.ents])

    for entity in entities:
        # Get sentences mentioning this entity
        entity_sents = [sent.text for sent in sentences if entity in sent.text]

        if not entity_sents:
            continue

        # Combine sentences mentioning entity for sentiment analysis
        combined_text = " ".join(entity_sents)

        # Apply your hybrid sentiment function here
        # For demo, using VADER sentiment:
        sentiment = get_vader_sentiment(combined_text)

        entity_sentiments[entity] = sentiment

    return entity_sentiments

# Example usage on a dataframe column:
df['Entity_Sentiments'] = df['short_summary'].apply(get_entity_sentiments)

# Now each row has a dict mapping entity names to their sentiment labels
print(df[['Tonality', 'Entity_Sentiments']])
```
