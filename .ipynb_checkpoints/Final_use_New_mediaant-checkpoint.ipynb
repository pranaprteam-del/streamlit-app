```python
import pandas as pd
import pickle
import time
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

# -------- Configuration --------
EXCEL_PATH = r"C:\Users\Sanket Shirke\MyModules\Scraping\New_mediaant_data\mediaant_publications.xlsx"
PICKLE_PATH = r"C:\Users\Sanket Shirke\MyModules\Scraping\New_mediaant_data\mediaant_editions.pkl"
GECKO_PATH = r"D:\Sanket Shirke\Downloads\geckodriver-v0.36.0-win64\geckodriver.exe"

SAVE_INTERVAL = 10 * 60  # Save every 10 minutes
HEADLESS = True
# --------------------------------

# -------- Setup Selenium Driver --------
options = Options()
options.headless = HEADLESS
service = Service(GECKO_PATH)
driver = webdriver.Firefox(service=service, options=options)
wait = WebDriverWait(driver, 20)

# -------- Scraping Function --------
def extract_all_card_links_with_load_more(url, driver, wait):
    try:
        driver.get(url)
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "h3.mediaCard_textStyles__2t7Nd")))

        last_card_count = 0
        retries = 0

        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

            try:
                load_more_button = driver.find_element(By.XPATH, "//button[contains(text(),'LOAD MORE')]")
                if load_more_button.is_displayed() and load_more_button.is_enabled():
                    driver.execute_script("arguments[0].click();", load_more_button)
                else:
                    break
            except Exception:
                break

            time.sleep(4)

            soup = BeautifulSoup(driver.page_source, 'html.parser')
            cards = soup.find_all('a', href=True)
            card_links = [a['href'] for a in cards if a.find('div', class_='mediaCard_cardContainer__1rich')]
            current_count = len(card_links)

            if current_count == last_card_count:
                retries += 1
                if retries >= 2:
                    print(f"Loaded {current_count} cards.")
                    break
            else:
                last_card_count = current_count
                retries = 0

        # Final extraction
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        a_tags = soup.find_all('a', href=True)

        links = []
        for a in a_tags:
            if a.find('div', class_='mediaCard_cardContainer__1rich'):
                href = a['href']
                if href.startswith('/'):
                    href = 'https://www.themediaant.com' + href
                links.append(href)

        return links

    except Exception as e:
        print(f"Error extracting from {url}: {str(e)}")
        return []

# -------- Load Data --------
try:
    df = pd.read_excel(EXCEL_PATH)
except Exception as e:
    print(f"Error reading Excel file: {str(e)}")
    driver.quit()
    raise

if 'url_edition' not in df.columns:
    df['url_edition'] = None

# -------- Scraping Loop --------
start_time = time.time()
last_save_time = start_time

try:
    for i, row in df.iterrows():
        pub = row.get('Publications', 'Unknown')
        url = row.get('url', '')

        if not url or not isinstance(url, str):
            print(f"[{i}] Invalid or missing URL for {pub}. Skipping.")
            continue

        if isinstance(row['url_edition'], list) and len(row['url_edition']) > 0:
            print(f"[{i}] Skipping already processed: {pub}")
            continue

        print(f"[{i}] Processing: {pub} - {url}")

        try:
            links = extract_all_card_links_with_load_more(url, driver, wait)
            df.at[i, 'url_edition'] = links
        except Exception as e:
            print(f"[{i}] Error processing {pub}: {str(e)}")
            continue

        # Save every SAVE_INTERVAL seconds
        current_time = time.time()
        if current_time - last_save_time > SAVE_INTERVAL:
            print(f"[{i}] Saving progress...")
            df_expanded = df.explode('url_edition').reset_index(drop=True)
            df_expanded.to_pickle(PICKLE_PATH)
            last_save_time = current_time

except KeyboardInterrupt:
    print("Interrupted by user. Saving progress...")

finally:
    # Final save
    df_expanded = df.explode('url_edition').reset_index(drop=True)
    df_expanded.to_pickle(PICKLE_PATH)
    driver.quit()
    print("Driver closed and data saved.")
```

```python
import pandas as pd
import pickle
import time
import os
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

# -------- Configuration --------
EXCEL_PATH = r"C:\Users\Sanket Shirke\MyModules\Scraping\New_mediaant_data\mediaant_publications.xlsx"
PICKLE_PATH = r"C:\Users\Sanket Shirke\MyModules\Scraping\New_mediaant_data\mediaant_editions.pkl"
GECKO_PATH = r"D:\Sanket Shirke\Downloads\geckodriver-v0.36.0-win64\geckodriver.exe"

SAVE_INTERVAL = 10 * 60  # Save every 10 minutes
HEADLESS = True
# --------------------------------

# -------- Setup Selenium Driver --------
options = Options()
options.headless = HEADLESS
service = Service(GECKO_PATH)
driver = webdriver.Firefox(service=service, options=options)
wait = WebDriverWait(driver, 20)

# -------- Scraping Function --------
def extract_all_card_links_with_load_more(url, driver, wait):
    try:
        driver.get(url)
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "h3.mediaCard_textStyles__2t7Nd")))

        last_card_count = 0
        retries = 0

        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

            try:
                load_more_button = driver.find_element(By.XPATH, "//button[contains(text(),'LOAD MORE')]")
                if load_more_button.is_displayed() and load_more_button.is_enabled():
                    driver.execute_script("arguments[0].click();", load_more_button)
                else:
                    break
            except Exception:
                break

            time.sleep(4)

            soup = BeautifulSoup(driver.page_source, 'html.parser')
            cards = soup.find_all('div', class_='mediaCard_cardContainer__1rich')
            current_count = len(cards)

            if current_count == last_card_count:
                retries += 1
                if retries >= 2:
                    print(f"Loaded {current_count} cards.")
                    break
            else:
                last_card_count = current_count
                retries = 0

        # Final extraction
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        cards = soup.find_all('div', class_='mediaCard_cardContainer__1rich')

        results = []
        for card in cards:
            parent_a = card.find_parent('a', href=True)
            if parent_a:
                href = parent_a['href']
                if href.startswith('/'):
                    href = 'https://www.themediaant.com' + href

                title_tag = card.find('h3')
                p_tag = card.find('p')

                title = title_tag.get_text(strip=True) if title_tag else None
                circulation = p_tag.get_text(strip=True) if p_tag else None

                results.append({
                    'url': href,
                    'title': title,
                    'circulation': circulation
                })

        return results

    except Exception as e:
        print(f"Error extracting from {url}: {str(e)}")
        return []

# -------- Load Data (prefer resume from pickle) --------
if os.path.exists(PICKLE_PATH):
    print("Loading existing progress from pickle...")
    df = pd.read_pickle(PICKLE_PATH)
else:
    print("Starting fresh from Excel...")
    try:
        df = pd.read_excel(EXCEL_PATH)
    except Exception as e:
        print(f"Error reading Excel file: {str(e)}")
        driver.quit()
        raise

# Ensure column exists
if 'card_data' not in df.columns:
    df['card_data'] = None

# -------- Scraping Loop --------
start_time = time.time()
last_save_time = start_time

try:
    for i, row in df.iterrows():
        pub = row.get('Publications', 'Unknown')
        url = row.get('url', '')

        if not url or not isinstance(url, str):
            print(f"[{i}] Invalid or missing URL for {pub}. Skipping.")
            continue

        if isinstance(row['card_data'], list) and len(row['card_data']) > 0:
            print(f"[{i}] Skipping already processed: {pub}")
            continue

        print(f"[{i}] Processing: {pub} - {url}")

        try:
            card_data = extract_all_card_links_with_load_more(url, driver, wait)
            df.at[i, 'card_data'] = card_data
        except Exception as e:
            print(f"[{i}] Error processing {pub}: {str(e)}")
            continue

        # Save every SAVE_INTERVAL seconds
        current_time = time.time()
        if current_time - last_save_time > SAVE_INTERVAL:
            print(f"[{i}] Saving progress...")
            df.to_pickle(PICKLE_PATH)
            last_save_time = current_time

except KeyboardInterrupt:
    print("Interrupted by user. Saving progress...")

finally:
    # Final save
    df.to_pickle(PICKLE_PATH)
    driver.quit()
    print("Driver closed and data saved.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import pandas as pd
import time
import os
import re

#  File paths
PICKLE_PATH = r'C:\Users\Sanket Shirke\MyModules\Scraping\New_mediaant_data\mediaant_editions_final.pkl'
OUTPUT_PATH = r'C:\Users\Sanket Shirke\MyModules\Scraping\New_mediaant_data\updated_mediaant_editions_final.pkl'

#  Load existing progress or raise error
if os.path.exists(PICKLE_PATH):
    df = pd.read_pickle(PICKLE_PATH)
    print(f" Loaded {len(df)} rows from pickle.")
else:
    raise FileNotFoundError(" Pickle file not found.")

#  Ensure ad_rate column exists
if 'ad_rate' not in df.columns:
    df['ad_rate'] = None

#  Clean up ad_rate column (optional, adjust logic if needed)
df['ad_rate'] = df['ad_rate'].apply(
    lambda x: int(re.sub(r"[^\d]", "", str(x))) if pd.notna(x) and str(x).strip() not in ["", "None", "nan"] and str(x).isdigit() else x
)

#  Setup Selenium (Firefox headless)
options = Options()
options.headless = True
options.binary_location = r"C:\Program Files\Mozilla Firefox\firefox.exe"
service = Service(r"D:\Sanket Shirke\Downloads\geckodriver-v0.36.0-win64\geckodriver.exe")
driver = webdriver.Firefox(service=service, options=options)
wait = WebDriverWait(driver, 15)

#  Scraping logic
def scrape_ad_rate(url):
    try:
        driver.get(url)
        wait.until(EC.presence_of_element_located((By.TAG_NAME, 'body')))
        soup = BeautifulSoup(driver.page_source, "html.parser")

        cards = soup.find_all("div", class_="MuiCardContent-root")

        #  Only consider "Custom Sized Ads"
        for card in cards:
            heading = card.find("h3")
            if heading and "Custom Sized Ads" in heading.get_text(strip=True):
                price_tag = card.find("p", class_=re.compile("mediaOption_rates"))
                if price_tag:
                    raw_rate = price_tag.get_text(strip=True)

                    #  Check for special labels
                    if "Per Insert" in raw_rate:
                        return "Per Insert"
                    if "On Request" in raw_rate:
                        return "On Request"

                    #  Extract numeric part
                    numeric_rate = re.sub(r"[^\d]", "", raw_rate)
                    if numeric_rate:
                        return int(numeric_rate)
                    else:
                        return None  # no valid number

        # If "Custom Sized Ads" card not found
        return "Not Found"

    except Exception as e:
        print(f" Error on {url}: {str(e)[:100]}")
        return None

#  Loop and scrape missing entries with safe saving
try:
    row_counter = 0
    for idx, row in df.iterrows():
        if pd.notna(row.get('ad_rate')) and str(row['ad_rate']).strip() not in ["", "None", "nan"]:
            continue  # Already scraped

        url = row.get('edition_url')
        if not url:
            continue

        print(f" Scraping row {idx + 1}: {url}")
        ad_rate = scrape_ad_rate(url)
        df.at[idx, 'ad_rate'] = ad_rate

        row_counter += 1

        #  Save progress every 100 rows
        if row_counter % 100 == 0:
            df.to_pickle(PICKLE_PATH)
            print(f" Progress saved at row {idx + 1}")
        
        time.sleep(3)  # polite delay

except KeyboardInterrupt:
    print("\n Keyboard Interrupt detected! Saving progress...")
    df.to_pickle(PICKLE_PATH)

except Exception as e:
    print(f"\n Unexpected error: {e}. Saving progress before exit...")
    df.to_pickle(PICKLE_PATH)
    raise  # optionally re-raise after saving

finally:
    # Final save and cleanup
    print(" Final save and closing browser...")
    df.to_pickle(PICKLE_PATH)
    driver.quit()

#  Convert numeric-looking values to int, skip strings like "On Request" or "Not Found"
columns_to_convert = ['edition_circulation', 'ad_rate']
for col in columns_to_convert:
    if col in df.columns:
        df[col] = df[col].apply(
            lambda x: int(x) if isinstance(x, (int, float)) and not pd.isna(x) else x
        )

#  Final export after all scraping
df.to_pickle(OUTPUT_PATH)
df.to_excel(OUTPUT_PATH.replace(".pkl", ".xlsx"), index=False)
print(" Scraping complete. Results saved to Excel and Pickle.")
```
